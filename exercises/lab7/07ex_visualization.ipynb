with Pandas and create scatter plots with all possible combinations of the following features:

Are these features correlated?

````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#!wget https://www.dropbox.com/s/aamg1apjhclecka/regression_generated.csv

file_name = "./data/regression_generated.csv"
data = pd.read_csv(file_name)
display(data)

sns.pairplot(data[["features_1", "features_2", "features_3"]])
plt.tight_layout()
plt.show()

# Check of the correlation of features 1
corr12 = np.corrcoef(data["features_1"], data["features_2"])[0,1]
corr13 = np.corrcoef(data["features_1"], data["features_3"])[0,1]
corr23 = np.corrcoef(data["features_2"], data["features_3"])[0,1]

print(f"Correlation of feature 1 and 2: {corr12}")
print(f"Correlation of feature 2 and 3: {corr13}")
print(f"Correlation of feature 2 and 3: {corr23}")
````

ANSWER: As can be seen from the correlation values of the 3 possible feature combinations, the features are not correlated

3\. **Profile plot**

Produce a profile plot from a scatter plot.
* Download the following pickle file:

* Inspect the dataset, you'll find two variables (features)
* Convert the content to a Pandas Dataframe
* Clean the sample by selecting the entries (rows) with the absolute values of the variable "residual" smaller than 2
* Plot a Seaborn `jointplot` of "residuals" versus "distances", and use seaborn to display a linear regression. 

Comment on the correlation between these variables.

* Create manually (without using seaborn) the profile histogram for the "distance" variable; choose an appropriate binning.
* Obtain 3 numpy arrays:
  * `x`, the array of bin centers of the profile histogram of the "distance" variable
  * `y`, the mean values of the "residuals", estimated in slices (bins) of "distance"
  * `err_y`, the standard deviation of the of the "residuals", estimated in slices (bins) of "distance"
* Plot the profile plot on top of the scatter plot

````python
#!wget https://www.dropbox.com/s/3uqleyc3wyz52tr/residuals_261.pkl -P data/

file_name = './data/residuals_261.pkl'
data = pd.read_pickle(file_name).item()
    
# Convert the content to a Pandas Dataframe
dframe = pd.DataFrame(data, columns = ['residuals','distances'])
#display(dframe)

# Selecting the entries with the absolute values of the variable "residual" smaller than 2
dframe = dframe[abs(dframe['residuals']) < 2]
#display(dframe)

# Plot a Seaborn `jointplot` of "residuals" versus "distances"
plot_res_dis = sns.jointplot(x = 'distances', y = 'residuals', data = dframe, kind = "reg", joint_kws = {'line_kws':{'color':'red', 'label':'Linear Regression'}})
plot_res_dis.ax_joint.legend()
plt.show()

corr_dis_res = np.corrcoef(dframe["distances"], dframe["residuals"])[0,1]
print(f"Correlation of distances and features: {corr_dis_res}")
````

ANSWER: As can be seen from the jointplot or from the correlation value, the feature "distances" and "features" are not correlated. Indeed the slope of the regression line (in red) is very close to zero.

````python
# Histogram of distances
num_bin = 20
(n, bins, patches) = plt.hist(dframe['distances'], num_bin, range=(0,20), alpha=1, histtype='bar', edgecolor='black')
plt.title("Distances")
plt.show()

# Array of bin centers of the profile histogram of the "distance" variable
x = 0.5*(bins[1:]+bins[:-1])
# Mean values of the "residuals" (estimated in slices (bins) of "distance)
y = [dframe["residuals"][(dframe['distances'] >= bins[i]) & (dframe['distances'] <= bins[i+1])].mean() for i in range(num_bin)]
# Standard deviation of the of the "residuals" (estimated in slices (bins) of "distance)
err_y = [dframe["residuals"][(dframe['distances'] > bins[i]) & (dframe['distances'] < bins[i+1])].std() for i in range(num_bin)]

print(f"\nArray of bin centers of the profile histogram of the distance variable, \nx: {x}")
print(f"\nMean values of the residuals (estimated in slices (bins) of distance), \ny: {y}")
print(f"\nThe standard deviation of the of the residuals (estimated in slices (bins) of distance), \nerr_y: {err_y}")

# Plot the profile plot on top of the scatter plot
plot_res_dis2 = sns.jointplot(x = 'distances', y = 'residuals', data = dframe, kind = "reg", joint_kws = {'line_kws':{'label':'Linear Regression', 'color':'red'}})
plot_res_dis2.ax_joint.legend() # print the legend
plt.sca(plot_res_dis2.ax_joint)
plt.errorbar(x, y , yerr = err_y, marker = 'o', color = 'orange', xlolims = 35, label = "Profile")
plt.legend()
plt.show()
````

4\. **Kernel Density Estimate**

Produce a KDE for a given distribution (by hand, not using seaborn):

* Fill a numpy array `x` of length N (with $N=\mathcal{O}(100)$) with a variable normally distributed, with a given mean and standard deviation
* Fill an histogram in pyplot taking proper care of the aesthetic:
   * use a meaningful number of bins
   * set a proper y axis label
   * set proper value of y axis major ticks labels (e.g. you want to display only integer labels)
   * display the histograms as data points with errors (the error being the poisson uncertainty)
* For every element of `x`, create a gaussian with the mean corresponding to the element value and the standard deviation as a parameter that can be tuned. The standard deviation default value should be:
$$ 1.06 * x.std() * x.size ^{-\frac{1}{5}} $$
you can use the scipy function `stats.norm()` for that.
* In a separate plot (to be placed beside the original histogram), plot all the gaussian functions so obtained
* Sum (with `np.sum()`) all the gaussian functions and normalize the result such that the integral matches the integral of the original histogram. For that you could use the `scipy.integrate.trapz()` method. Superimpose the normalized sum of all gaussians to the first histogram.


````python
from scipy.stats import norm
from scipy.integrate import trapz

N = 100
mean = 0
std = 3
x = np.random.normal(loc=mean, scale=std, size=N)

# Definition of a figure with 3 subfigures
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(14, 4))

#Plotting the Histogram
bins = ax1.hist(x, bins=20, color='yellow',edgecolor='black')

error = np.sqrt(bins[0])
bin_center = 0.5*(bins[1][:-1] + bins[1][1:])
ax1.errorbar(bin_center, bins[0], yerr=error, fmt='r.')

ax1.set_title('Histogram')
ax1.set_ylabel('Entries')

# Standard deviation
std = 1.06*x.std()*(x.size**(-0.2))
x_g = np.linspace(np.min(x), np.max(x), N)

# Sum all the gaussian functions
gfunctions = []
for sample in x:
    gaussian = norm.pdf(x_g, sample, std)
    gfunctions.append(gaussian)
    ax2.plot(x_g, gaussian)

ax2.set_title('Gaussian')
ax2.set_ylabel('Entries')

#KDE
sumgaus = np.sum(gfunctions, 0) #Sum of all Gaussian function
sumgaus_area = trapz(sumgaus, x_g)
hist_area = sum(np.diff(bins[1])*bins[0])
k_norm = sumgaus_area/hist_area
sumgaus_norm = sumgaus/k_norm # Normalization

ax3.plot(x_g, sumgaus_norm, color="red")
ax3.hist(x, bins=20, color='yellow',edgecolor='black')

ax3.set_title('KDE')
ax3.set_ylabel('Entries')

# Plot the resulting graphs
plt.show()
````
