```python
#Number representation
#03ex_introduction_es01

def Converter(inpt, outTyp):
    try: #Check if the input is binary
        val = int(inpt,2) #binary if correct
        if outTyp == "binary":
            return val
        elif outTyp == "hexadecimal":
            tmp = val
            h = hex(tmp)
            return h
        else:
            return val
    except: 
        #Not binary
        #decimal or hexadecimal
        try: #hexadecimal check
            val = int(inpt,16)
            if outTyp == "binary":
                b = bin(val)
                return b
            elif outTyp == "hexadecimal":
                return val
            else:
                return val
        except: #decimal
            if outTyp == "binary":
                b = bin(inpt)
                return b
            elif outTyp == "hexadecimal":
                return hex(inpt)
            else:
                return inpt
            
Converter(89342,"hexadecimal")
```

```python
#32-bit floating point number
##03ex_introduction_es02

def convertBinaryToFloat(e, flag, f):
    exponent = e[::-1]
    sum_of_e = 0
    sum_of_f = 0
    for i in range(0,len(exponent)):
        sum_of_e += (int(exponent[i]) * (2**(i)))
    for i in range(0,len(f)):
        sum_of_f += (int(f[i]) * (2**(-(i+1))))
    result = (flag * ( 1 + sum_of_f)) * (2**(sum_of_e - 127))
    print(result)

binary = "0000001111100000000000001100111100"
              
s, e, f = binary[0], binary[1:9], binary[9:]

if s == '0':
    convertBinaryToFloat(e,1,f)
else:
    convertBinaryToFloat(e,-1,f)     

    
    
```

```python
#The derivative
#03ex_introduction_es05

import math

def quadrat_func(a,b,c):
    pos_root = (((-b) + math.sqrt((b**2) - ((4*a)*c))) / (2*a))
    neg_root = (((-b) - math.sqrt((b**2) - ((4*a)*c))) / (2*a))
    return pos_root, neg_root

pos_root, neg_root = quadrat_func(0.001,1000,0.001)

print("first root -> " + str(pos_root) + " and second root -> " + str(neg_root))


def quadrat_modified_func(a,b,c):
    delta = math.sqrt((b**2) - ((4*a)*c))
    root1 = ((((-b) + delta) * ((-b) + delta)) / ((2*a)*((-b) + delta)))
    root2 = ((((-b) + delta) * ((-b) - delta)) / ((2*a)*((-b) - delta)))
    root3 = ((((-b) - delta) * ((-b) + delta)) / ((2*a)*((-b) + delta)))
    root4 = ((((-b) - delta) * ((-b) - delta)) / ((2*a)*((-b) - delta)))
    return root1, root2, root3, root4
    
root1, root2, root3, root4 = quadrat_modified_func(0.001,1000,0.001)
print("first root -> " + str(root1) + ", second root -> " + str(root2) +
     ", third root -> " + str(root3) + " and fourth root -> " + str(root4))

```

```python
#The derivative
#03ex_introduction_es06

def LimitDer(s):
    x = 1
    def f(x):
        return x * (x-1)
    return (f(x+s) - f(x)) / s
der = LimitDer(10**-2)
print(der)
```
