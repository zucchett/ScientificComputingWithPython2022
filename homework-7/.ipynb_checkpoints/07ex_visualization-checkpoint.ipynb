2\. **Color-coded scatter plot**

Produce a scatter plot from a dataset with two categories.

* Write a function that generates a 2D dataset consisting of 2 categories. Each category should distribute as a 2D gaussian with a given mean and standard deviation. Set different values of the mean and standard deviation between the two samples.
* Display the dataset in a scatter plot marking the two categories with different marker colors.

An example is given below:

````python
from IPython.display import Image
Image('images/two_categories_scatter_plot.png')
````

````python
def dataset(mean, std):
    x = 2
    data = []
    for i in range(x):
        data.append(np.random.normal(mean, std, 100))
    return data

mean1 = 4
std1 = 4
data1 = dataset(mean1, std1)
mean2 = 20
std2 = 10
data2 = dataset(mean2, std2)


fig = plt.figure()
plt.scatter(data1[0], data1[1], c='g', edgecolor = 'black', label = 'mean = 4, std = 4')
plt.scatter(data2[0], data2[1], c='b', edgecolor = 'black', label = 'mean = 20, std = 10')
plt.legend(loc='upper right')
plt.show()
````

````python
import seaborn as sns

data = dict(np.load('data/residuals_261.pkl', allow_pickle = True).item())
df = pd.DataFrame(data)
clean_data = df[np.absolute(df['residuals']) < 2]
sns.jointplot(x = 'distances', y = 'residuals', data = clean_data, kind = 'reg')

nrnd = clean_data['distances'] 
fig, ax = plt.subplots()
h, bins, _ = plt.hist(nrnd, bins=20, density=True) 
nrnd.plot.kde(ax=ax, legend=False)
ax.set_title('Residuals')
ax.set_xlabel('Distance')
ax.set_ylabel('Probability')
fig.tight_layout()
````
