# **Exercise 1**

```python
hexa ='A''B''C''D''E''F''a''b''c''d''e''f'
deci ='2''3''4''5''6''7''8''9'
bins = '0''1'
input1 = input()

def IsBins(inpt):
  for char in inpt:
    if char in bins:
     return True

def Isdecis(inpt):
  for char in inpt:
    if char in deci:
      return True

def IsHexs(inpt):
    for char in inpt:
      if char in hexa:
        return True

if IsHexs(input1):
  input_3 = int(input1,16)
  print("Your input is Hexadecimal.")
elif Isdecis(input1):
  input_3 = int(input1) 
  print("Your input is Decimal.")
elif IsBins(input1):
  input_3 = int(input1,2)
  print("Your input is Binary.")

input2= input("Convert it to: \n 1)Binary \n 2)Decimal \n 3)Hexadecimal")

if input2 == '1':
  print(bin(input_3))
elif input2 == '2':
  print(int(input_3))
elif input2 == '3':
  print(hex(input_3))
```

# **Exercise 2**

```python
def Convert(strs):
  power = -1
  mantissa = 0

  for i in strs:
    mantissa += (int(i) * pow(2, power))
    power -= 1    
  return (mantissa + 1)


ieee32 = input("Write the 32 bit binary string: ")
sign = int(ieee32[0])
bias = int(ieee32[1 : 9], 2)
unbias = bias - 127

mantstr = ieee32[9 : ]
mantint = Convert(mantstr)
realnumber = pow(-1, sign) * mantint * pow(2, unbias)

print("Float representation",realnumber)
```
