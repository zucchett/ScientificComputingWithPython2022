### 2. JSON files

```python
# Uncomment to download the file:
#!wget https://www.dropbox.com/s/sz5klcdpckc39hd/user_data.json -P ./data/
```

```python
with open("data/user_data.json") as json_file:
    cc_data = json.load(json_file)
```

```python
cc_data_csv = open("data/user_data_AmericanExpress.csv","w")
csv_writer = csv.writer(cc_data_csv)

filtered = list(filter(lambda x: x['CreditCardType'] == 'American Express', cc_data))
count=0
for card in filtered:
    if count==0:
        header = card.keys()
        csv_writer.writerow(header)
        count+=1
    csv_writer.writerow(card.values())

cc_data_csv.close()
```

### 4. Reading the credit card numbers

```python
# Uncomment to download the file:
#!wget https://www.dropbox.com/s/8m0syw2tkul3dap/credit_card.dat -P ./data/
```

```python
file = open("data/credit_card.dat")
lines = file.readlines()
file.close()

for line in lines: 
    splitted = [line[i:i+6] for i in range(0,len(line)-6,6)]
    dec = [int(x,2) for x in splitted]
    characters = [chr(x) for x in dec]
    cardnumber = "".join(characters)
    print(cardnumber)
```

### 5. Write data to a binary file

```python
# Uncomment to download the file:
#!wget https://www.dropbox.com/s/xvjzaxzz3ysphme/data_000637.txt -P ./data/
```

```python
data = pd.read_csv("data/data_000637.txt")
df = data[:10]

with open('data/bin_file.dat','w') as f:
    for i in range(10):
        word = ""
        word+='{0:02b}'.format(int(df.iloc[i][0]) & 0x3)
        word+='{0:04b}'.format(int(df.iloc[i][1]) & 0xF)
        word+='{0:09b}'.format(int(df.iloc[i][2]) & 0x1FF)
        word+='{0:32b}'.format(int(df.iloc[i][3]) & 0xFFFFFFFF)
        word+='{0:12b}'.format(int(df.iloc[i][4]) & 0xFFF)
        word+='{0:05b}'.format(int(df.iloc[i][5]) & 0x1F)
        
        f.write(word+"\n")
```

```python
short_file = open("data/data_000637_short.txt","w")
og_file = open("data/data_000637.txt","r")
og_file.readline()
for i in range(10):
    short_file.write(og_file.readline())
short_file.close()
og_file.close()
```
