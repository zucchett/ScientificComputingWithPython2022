Produce a scatter plot from a dataset with two categories.

````python
# Write a function that generates a 2D dataset consisting of 2 categories. 
# Each category should distribute as a 2D gaussian with a given mean and standard deviation. 
# Set different values of the mean and standard deviation between the two samples.
n_samples = 1000
data_1 = np.random.multivariate_normal((0,0), [[0.7, 0],[0, 0.7]], (n_samples,))
data_2 = np.random.multivariate_normal((1,1), [[0.2, 0],[0, 0.2]], n_samples)

# Display the dataset in a scatter plot 
# marking the two categories with different marker colors.

plt.scatter(data_1[:,0],data_1[:,1], c='y', label='data_1 mean (0,0)')
plt.scatter(data_2[:,0],data_2[:,1], c='purple', label='data_2 mean (1,1)')
plt.title('Color-coded scatter plot')
````

````python
data = pd.DataFrame( np.load("data/residuals_261.pkl", allow_pickle=True).item())

data = data[abs(data['residuals'])<2]

ln = stats.linregress(data['distances'],data['residuals'])

f = sns.jointplot(data['distances'],data['residuals'],kind='reg')

plt.sca(f.ax_joint)
x = np.linspace(0,20,20)

bin_idx = np.digitize(data['distances'], x) 
y = [data[bin_idx == i].mean()[0] for i in range(0, len(x))]
err_y = [data[bin_idx == i].std()[0] for i in range(0, len(x))]

plt.errorbar(x,y,yerr=err_y, color='aquamarine')
````

4\. **Kernel Density Estimate**

Produce a KDE for a given distribution (by hand, not using seaborn):

* Fill a numpy array `x` of length N (with $N=\mathcal{O}(100)$) with a variable normally distributed, with a given mean and standard deviation
* Fill an histogram in pyplot taking proper care of the aesthetic:
   * use a meaningful number of bins
   * set a proper y axis label
   * set proper value of y axis major ticks labels (e.g. you want to display only integer labels)
   * display the histograms as data points with errors (the error being the poisson uncertainty)
* For every element of `x`, create a gaussian with the mean corresponding to the element value and the standard deviation as a parameter that can be tuned. The standard deviation default value should be:
$$ 1.06 * x.std() * x.size ^{-\frac{1}{5}} $$
you can use the scipy function `stats.norm()` for that.
* In a separate plot (to be placed beside the original histogram), plot all the gaussian functions so obtained
* Sum (with `np.sum()`) all the gaussian functions and normalize the result such that the integral matches the integral of the original histogram. For that you could use the `scipy.integrate.trapz()` method. Superimpose the normalized sum of all gaussians to the first histogram.


````python
import scipy as sp
x= np.random.normal(loc=0, scale=1, size=1000)

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
y, var, bins = ax1.hist(x,bins=20,histtype='step')

ax1.errorbar(0.5*(var[1:] + var[:-1]), y, yerr = y**0.5, label = 'Error')
ax1.set_title('KDE of X')
ax1.set_ylabel('Entries')
ax1.set_yticks(ticks=np.linspace(0,y.max(),10,dtype=int))


tem = []
for i in x:
    ls = np.linspace(-5, +5, num=1000)
    rv = stats.norm(i, 1.06 * x.std() * x.size**(-1/5))
    tem.append(rv.pdf(ls))
    ax2.plot(ls, tem[-1])
ax2.set_title('Gaussians of x values')

ax1.plot(ls, np.sum(tem, axis=0)/1000*(sp.integrate.trapz(y,x=var[1:])),label='KDE')
````
