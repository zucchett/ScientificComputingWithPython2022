```python
import numpy as np
import pandas as pd
import random 
import csv
import itertools
```

```python
f = open("data_int.txt", "w")
listofnums = []
n = random.randint(1,1000)
for i in range(0,n):
  listofnums.append(random.randint(0,10000))
f.write(str(listofnums)) 
f.close()

!cat data_int.txt
```

```python
mat = np.random.rand(5, 5)
x = open("data_float.txt", "w")
x.write(str(mat)) 
x.close()
!cat data_float.txt
```

```python
with open("data_float.txt", 'r') as infile, open("data_float.csv", 'w') as outfile:
        stripped = (line.strip() for line in infile)
        lines = (line.split(",") for line in stripped if line)
        writer = csv.writer(outfile)
        writer.writerows(lines)

!cat data_float.csv
```
