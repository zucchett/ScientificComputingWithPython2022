```python
###########Esercizio 1
import numpy as np
import math
from numpy import random
m = np.arange(12).reshape((3,4))
tMean=m.mean()
colMean=m.mean(0)
rowMean=m.mean(1)
print("the mean for each row is : ",rowMean,"\n")
print("the mean for each column is ",colMean,'\n')
print("the mean of total array is : ",tMean,"\n")
#another way to compute  total mean
ta=sum(sum(m[:,:]))/12
print("using another way ",ta,"\n")
############# Esercizio 2
u = np.array([1, 3, 5, 7])
v = np.array([2, 4, 6, 8])
outerP1=np.outer(u,v)
print("outer product using oter method : ","\n",outerP1,"\n")
o=[]
for i in u :
    for j in v :
        o.append(i*j)
o=np.array(o)
print("outer product using two nested for loops: \n",o.reshape(4,4),"\n")

###################################################### Esercizio 3
z=np.random.uniform(0.0,3.0,(10,6))
print(" before  \n",z)
z[z<0.3]=0

print("after \n",z)
################################## Esercizio 4

k=np.linspace(0,2*(math.pi),100)
print("linspace : \n", k)
result= k[::10]
print("\n Every 10th elements : \n",result)
#######
print("reversed : \n",k[100:0:-1])
#######
mask=(abs(np.sin(k)-np.cos(k))>0.3)
print("the mask is : \n",mask)
print(" element which sin - cos more than 0.3 are \n ",k[mask])
```
