4\. **Reading the credit card numbers**

```python
#!wget https://www.dropbox.com/s/8m0syw2tkul3dap/credit_card.dat
import math as m

with open('credit_card.dat', 'rb') as file:
    file_content = file.readlines()
    #print(file_content)
    allcarts =[]
    word_counter = 0
    word_size = 8 # size of the word in bytes
    total_digitnumber = 16+3 #size of total digit
    digit_length = 6          #sizeof each character with whitespace
    
    
    for row in file_content:
        #print(round(len(row)/digit_length))
        if(round(len(row)/digit_length) >= total_digitnumber):
            
            utf_string = row.decode("utf-8")  #decode the line into string
            #print("1:\n", utf_string, "\n")
            utf_string = utf_string[:-4]      #remove the last 4 bit : end of line
            #print("2:\n", utf_string, "\n")
            
            start_index = 0
            card_number=[]
            
            for i in range(total_digitnumber):
                end_index = start_index + digit_length
                digits_4 = utf_string[start_index:end_index]
                card_number.append(digits_4)
                start_index += digit_length    
                #print(digits_4 ,"\n")
        
            #print("card number: ", card_number, "\n")

            credit_card = ""
            for i in range(total_digitnumber):
                credit_card += chr(int(card_number[i],2))
            
            allcarts.append(credit_card)
            
    print("Number of all the Credit Card is :\n") 
    for i in allcarts:
        print(i, "\n")
        
        
```

5\. **Write data to a binary file**

```python
import pandas as pd
from struct import pack, unpack


#a) Packing-----------
binary_file= open('data/data_000637.dat', 'wb')
file_name= "data/data_000637.txt"

data = pd.read_csv(file_name, nrows =10, skiprows=range(1,1))
print("data: \n", data, "\n")
print("Binary: ")
for i in range(len(data)):
    h = data.loc[data.index[i], 'HEAD']
    f = data.loc[data.index[i], 'FPGA']
    t = data.loc[data.index[i], 'TDC_CHANNEL']
    o = data.loc[data.index[i], 'ORBIT_CNT']
    b = data.loc[data.index[i], 'BX_COUNTER']
    td = data.loc[data.index[i], 'TDC_MEAS']
    
    word  = (h & 0x3) << 62
    word += (f & 0xF) << 58
    word += (t & 0x1FF) << 49
    word += (o & 0xFFFFFFFF) << 17
    word += (b & 0xFFF) << 5
    word += (td & 0x1F) << 0
    packed = pack('<q', word)
    print(packed)
    binary_file.write(packed)
    
binary_file.close()

#b) Unpacking-----------
with open('data/data_000637.dat', 'rb') as file:
    file_content = file.read()
    word_counter = 0
    word_size = 8 # size of the word in bytes
    for i in range(0, len(file_content), word_size):
        word_counter += 1
        if word_counter > 10: break
        wd = unpack('<q', file_content[i : i + word_size])[0] # get an 8-byte word
        
        head     = (wd >> 62) & 0x3
        fpga     = (wd >> 58) & 0xF
        tdc_chan = (wd >> 49) & 0x1FF
        orb_cnt  = (wd >> 17) & 0xFFFFFFFF
        bx       = (wd >> 5 ) & 0xFFF
        tdc_meas = (wd >> 0 ) & 0x1F
        print('\n{0}\t{1}\t{2}\t{3}\t{4}\t{5}'.format(head, fpga, tdc_chan, orb_cnt, bx, tdc_meas))
        
#c)
print("\nThe size of binary and txt file are respectfully 80 and 312 bytes which shows\n" +
       "saving data in binary format is much more better than .txt due to the less memory consumption")
```
