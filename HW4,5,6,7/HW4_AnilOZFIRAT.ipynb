1\. **Reductions**

Find the total mean, and the mean for each row and column of the following matrix:

```python

```

2\. **Outer product**

Find the outer product of the following vectors:

```python
u = np.array([1, 3, 5, 7])
v = np.array([2, 4, 6, 8])
```

Use different methods to do this:

   1. Using the function `outer` in numpy
   2. Using a nested `for` loop or a list comprehension
   3. Using numpy broadcasting operations

````python
import numpy as np

u = np.linspace(1, 7, 4)
v = np.linspace(2, 8, 4)

print("u array : ",u)
print("v array : ",v)

# First method
output = np.outer(u, v)
print("\n1- Result of outer: \n",output,"\n")

# Second method

output2 = (np.array([(i*v) for i in u]))
print("2- Result of for :\n", output2)
print("\n")
#print("2- Result of for : \n",output2,"\n")


# Third Method
output3 = u.reshape(-1,1)*v
print("3- Result of boradcasting\n", output3)
````

5\. **Matrices**

Create a matrix that shows the 10 by 10 multiplication table.

 * Find the trace of the matrix
 * Extract the anti-diagonal matrix (this should be ```array([10, 18, 24, 28, 30, 30, 28, 24, 18, 10])```)
 * Extract the diagonal offset by 1 upwards (this should be ```array([ 2,  6, 12, 20, 30, 42, 56, 72, 90])```)

````python
import numpy as np

x = range(1,11) 
y = range(1,11)
mult_tab = np.asarray([[i*j for j in x] for i in y])
shape = (10,10)
print("10x10 table:\n",mult_tab.reshape(shape))
print("1- Trace of the matrix is", mult_tab.trace())
print("2- Anti-diagona matrix", np.fliplr(mult_tab).diagonal())
print("3- Diagonal offset by 1 upwards",mult_tab.diagonal(1))
````

6\. **Broadcasting**

Use broadcasting to create a grid of distances.

Route 66 crosses the following cities in the US: Chicago, Springfield, Saint-Louis, Tulsa, Oklahoma City, Amarillo, Santa Fe, Albuquerque, Flagstaff, Los Angeles.

The corresponding positions in miles are: 0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448

  * Build a 2D grid of distances among each city along Route 66
  * Convert the distances in km

````python
import numpy as np

c = ["Chicago","Springfield","Saint-Louis","Tulsa","Oklahoma City","Amarillo","Santa Fe","Albuquerque","Flagstaff","Los Angeles"]
d = np.array([0,198,303,736,871,1175,1475,1544,1913,2448])
G = np.zeros((d.size,d.size))
for i in range(d.size):
    for j in range(d.size):
        G[i,j] = abs(d[i]-d[j])

print("Distances among each city as Miles\n",G)
print("\nDistances among each city as KM\n",np.around(G*(1.609344)))
````

8\. **Diffusion using random walk**

Consider a simple random walk process: at each step in time, a walker jumps right or left (+1 or -1) with equal probability. The goal is to find the typical distance from the origin of many random walkers after a given amount of time.

*Hint*: create a 2D array where each row represents a walker, and each column represents a time step.

  * Take 1000 walkers and let them walk for 200 steps
  * Use `randint` to create a 2D array of size $walkers \times steps$ with values -1 or 1
  * Calculate the walking distances for each walker (e.g. by summing the elements in each row)
  * Take the square of the previously-obtained array (element-wise)
  * Compute the mean of the squared distances at each step (i.e. the mean along the columns)
  * **Optional**: plot the average distances ($\sqrt(distance^2)$) as a function of time (step)

````python
import numpy as np
import random
import matplotlib.pyplot as plt

y = []
x = (2 * np.random.randint(0,2,(1000,200))) -1                  # 1000 x 200 array
[y.append(abs(sum(x[i,:])))  for i in range(0,1000)]            # distances of walker (summing)
print("Distances of walker to center:\n",y)

y_sqr = np.multiply(y,y)                                        # sqauare of distances
print("Square of distances:\n",y_sqr)

x_sum = np.cumsum(x, axis=1)                                    # Compute the distances at each step
x_sqr = np.multiply(x_sum,x_sum)                                # sqaured distances at each step

mean = [np.mean(x_sqr[:,i]) for i in range(0,200)]              # mean squared distances along with each step
print("mean squared distances along with each step:\n",mean)

walkers = np.arange(0, len(mean))                     
plt.rcParams["figure.figsize"] = (5,5)
plt.plot(walkers,mean)                                          # average distance
plt.show()
````
