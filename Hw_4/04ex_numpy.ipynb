1\. **Reductions**

Find the total mean, and the mean for each row and column of the following matrix:

```python
m = np.arange(12).reshape((3,4))
```

````python
m = np.arange(12).reshape((3,4))
row_mean=m.mean(axis=0)
colm_mean=m.mean(axis=1)
print(m)

for i in range(m.shape[0]):
    colm=colm_mean[i]
    print("The mean of colm of %d is = " %(i+1),colm)

for i in range(m.shape[1]):
    row=row_mean[i]
    print("The mean of row of %d is = " %(i+1),row)

print("The total of mean is =",m.mean())
````

2\. **Outer product**

Find the outer product of the following vectors:

```python
u = np.array([1, 3, 5, 7])
v = np.array([2, 4, 6, 8])
```

Use different methods to do this:

   1. Using the function `outer` in numpy
   2. Using a nested `for` loop or a list comprehension
   3. Using numpy broadcasting operations

````python
#1-
u = np.array([1, 3, 5, 7])
v = np.array([2, 4, 6, 8])

np_outer=np.outer(u,v,out=None)
print("Outer produc by using numpy: \n",np_outer)


#2-

list_comp_outer=[(u[i]*v[j]) for i in range(len(u)) for j in range(len(v))]
comp_outer=np.array(list_comp_outer)
print ("\nOuter produc by using list compression: \n",comp_outer.reshape(4,4))

#3-

np_brodcasting= u[:, np.newaxis] * v
print("\nOuter produc by using numpy brodcasting: \n",np_brodcasting)
````

5\. **Matrices**

Create a matrix that shows the 10 by 10 multiplication table.

 * Find the trace of the matrix
 * Extract the anti-diagonal matrix (this should be ```array([10, 18, 24, 28, 30, 30, 28, 24, 18, 10])```)
 * Extract the diagonal offset by 1 upwards (this should be ```array([ 2,  6, 12, 20, 30, 42, 56, 72, 90])```)

````python
matrix = np.zeros((10,10))

for i in range(10):
    for j in range(10):
        matrix[i][j] = (i+1) * (j+1)
print("Matrix that shows the 10 by 10 multiplication table: \n",matrix)
print("\nTrace of matrix: ",np.trace(matrix))
print("\nAnti-diagonal offset by 1 upwards are :\n",np.diag(matrix[::-1], 0))
print("\nDiagonal offset by 1 upwards are :\n",np.diag(matrix, 1))
````

6\. **Broadcasting**

Use broadcasting to create a grid of distances.

Route 66 crosses the following cities in the US: Chicago, Springfield, Saint-Louis, Tulsa, Oklahoma City, Amarillo, Santa Fe, Albuquerque, Flagstaff, Los Angeles.

The corresponding positions in miles are: 0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448

  * Build a 2D grid of distances among each city along Route 66
  * Convert the distances in km

````python
miles = np.array([0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448])
miles_n = np.zeros((10,10))
km = np.zeros((10,10))

for i in range(len(miles_n)):
    miles_n[i] = abs(miles[i] - miles)
km = miles_n * 1.609344
print(miles_n,"\n")
print(km)
````

7\. **Prime numbers sieve**

Compute the prime numbers in the 0-N (start with N=99) range with a sieve (mask).

  * Constract a shape (N,) boolean array, which is the mask
  * Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element
  * Apply the mask to obtain an array of ordered prime numbers
  * Check the performances (with `timeit`); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

````python
def prime_num(N):
    primeNumberMask = np.array(np.arange(0,N+1),bool)
    primeNumberMask[:2] = False
    for i in range(2,N+1):
        for j in range(i+1,N+1):
            if((j)%(i)==0 and primeNumberMask[j]== True):
                primeNumberMask[j]=False
            
    return primeNumberMask

def sieve_of_erathosthenes(x):
    sieve=[]
    for i in range(x+1):
        i=True
        sieve.append(i)
    sieve[0], sieve[1] = False, False
    for i in range(2, int(x**0.5)+1):
        for j in range(2**i, x+1, i):
            sieve[j] = False
    primes=[]
    for i in range(2, x+1):
        if sieve[i] == True:
            primes.append(i)
    return primes


N=int(input("Please enter range of prime number whict will calculate: "))  
starttime = timeit.default_timer()
primeNumberMask=prime_num(N)
endtime = timeit.default_timer()

boolean=np.arange(0,N+1)


starttime_1 = timeit.default_timer()
sieve=sieve_of_erathosthenes(N)
endtime_1 = timeit.default_timer()
    
print("\n0 to N, prime numbers: \n",boolean[primeNumberMask])
print("\nMask array:\n",primeNumberMask)
print("\nChecked performance of scale of N as second: ",(endtime-starttime))
%timeit prime_num(N)
print("\nsieve result",sieve)
print("\nChecked performance of scale of N as second by usnig sieve of Eratosthenes:",(endtime_1-starttime_1))
%timeit sieve_of_erathosthenes(N)
````

8\. **Diffusion using random walk**

Consider a simple random walk process: at each step in time, a walker jumps right or left (+1 or -1) with equal probability. The goal is to find the typical distance from the origin of many random walkers after a given amount of time.

*Hint*: create a 2D array where each row represents a walker, and each column represents a time step.

  * Take 1000 walkers and let them walk for 200 steps
  * Use `randint` to create a 2D array of size $walkers \times steps$ with values -1 or 1
  * Calculate the walking distances for each walker (e.g. by summing the elements in each row)
  * Take the square of the previously-obtained array (element-wise)
  * Compute the mean of the squared distances at each step (i.e. the mean along the columns)
  * **Optional**: plot the average distances ($\sqrt(distance^2)$) as a function of time (step)

````python
walk = np.random.randint(0, 2, size=(1000*200))
for i in range(len(walk)):
    if walk[i] == 0:
        walk[i] = -1
walk = walk.reshape(1000,200)
print(" ->2D array of walkers x steps:\n",walk)
total_distances=walk.sum(axis=1).reshape(1000,1)
print("\n ->Total distances from starting point for each walkers:\n",total_distances)
distance_by_step=walk.cumsum(axis=1)
distance_by_step=distance_by_step.reshape(1000,200)
print("\n ->Distances for each step for each walkers:\n",distance_by_step)
square=np.square(distance_by_step).reshape(1000,200)
print("\n ->Square of the distances for each step for each walkers:\n",square)
mean_colmn=[np.mean(square[:,i]) for i in range(200)]
print("\n ->Mean of the squared distances at each step:\n",mean_colmn)

#####################-Optional-##############################
m = np.arange(0, len(mean_colmn))
plt.plot(m, mean_colmn)
plt.xlabel('Step')
plt.ylabel('Mean of square of the distances as column')
plt.title(' Average Distances ')
plt.show()
````
