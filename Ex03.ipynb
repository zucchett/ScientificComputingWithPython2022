#1. Number representation



```python
inp = input()

Hexx = 'A''B''C''D''E''F''a''b''c''d''e''f'
Dec = '2''3''4''5''6''7''8''9'

def determine(numb):
  
  outpt = "Binary"

  for ch in Hexx:
    if ch in numb:
      outpt = "Hexadecimal"
      break

  if outpt != "Hexadecimal":
    for ch in Dec:
      if ch in numb:
        outpt = "Decimal"
        break

  return outpt

inp_represent = determine(inp)

print("input type is:" , inp_represent)


if inp_represent == "Hexadecimal":
    int_inp = int(inp, 16)
elif inp_represent == "Binary":
    int_inp = int(inp, 2)
else:
    int_inp = int(inp)


act = input('What representation do you want? \n a. Binary \n b. Decimal \n c. Hexadecimal \n')

if act == 'a':
  print(bin(int_inp))

elif act == 'b':
  print(int_inp)

elif act == 'c':
  print(hex(int_inp))
```

#2. 32-bit floating point number

```python
def decimalRepresentation(mantissa):

	power_count = -1

	mantiInt = 0

	for i in mantissa:
		mantiInt += (int(i) * pow(2, power_count))
		power_count -= 1
		
	return (mantiInt + 1)


ieee = input()

sign_bit = int(ieee[0])

exponent_bias = int(ieee[1 : 9], 2)

exponent_unbias = exponent_bias - 127

mantissa_str = ieee[9 : ]
mantissa_int = decimalRepresentation(mantissa_str)
deci = pow(-1, sign_bit) * mantissa_int * pow(2, exponent_unbias)

print("The decimal representation of the 32-bit ieee is :",deci)
```

```python
import numpy as np

abc = input().split(' ')
a = float(abc[0])
b = float(abc[1])
c = float(abc[2])

def calc(aw, bw, cw):
  delta = bw**2 - 4*aw*cw

  x1 = ( (-bw) + np.sqrt(delta)) / (2*aw)
  x2 = ( (-bw) - np.sqrt(delta)) / (2*aw)

  anss = [x1.hex(), x2.hex()]

  return anss


print(calc(a, b, c))
```
