# **Exercise 2**

```python
u = np.array([1, 3, 5, 7])
v = np.array([2, 4, 6, 8])
import numpy as np
print("The solution with np.outer:\n",np.outer(u,v))


secondsolution = []
for i in u:
  row=[]
  for j in v:
    row.append(i*j)
  secondsolution.append(row)
print("\nThe solution with for loop:\n", secondsolution)


from numpy import array, newaxis
x = u[:,newaxis] * v
print("\nThe solution with  numpy broadcasting operations:\n",x)
```

```python
mat = np.linspace(0, 2*(np.pi), 100)
print("The matrix:\n",mat)

b = mat[9::10]
print("\nExtract every 10th element using the slice notation:\n",b)

c= mat[99:0:-1]
print("\nReverse the array using the slice notation:\n",c)

import math
d = []
for i in mat:
 
  if np.abs(math.sin(i)-math.cos(i))<0.1:
    d.append(i)
print("\nExtract elements where the absolute difference between the sin and cos functions evaluated for that element is <0.1:\n",d)

```

# **Exercise 7**

```python
def prime(x): 
  
  listofprimes = []
  for i in range(2,x+1):
    listofprimes.append(i)
  for i in range(2,x+1):
    for j in range(2,i):
      mask = (i % j == 0)
      if mask:
        if i in listofprimes:
          listofprimes.remove(i)

  arrayofprimes = np.array(listofprimes)
  print(arrayofprimes)

print("The prime number from 0 to 99:")
prime(100)

        
```

# **Exercise 8**

```python
from random import choice

walker = [[choice([-1, 1]) for i in range(200)] for j in range(1000)]
distanceofwalkers = []
lengthofwalker = len(walker)
for i in range(lengthofwalker):
  sumofwalker = sum(walker[i])
  distanceofwalkers.append(sumofwalker)
  print("The walker:", walker[i])
  print("The walker %d total distance: %d ." %(i, sumofwalker),"\n")
```

```python
squaredofsteps=[]
for i in distanceofwalkers:
  squaredofsteps.append(i**2)

for j in range(1000):
  print("The square distance for walker", " ","%2.5e"%(j+1),"is","\t\t", "|%2.5e"%squaredofsteps[j], "|")
```

```python
print("The mean of squared distances: ", sum(squaredofsteps)/1000)
```
